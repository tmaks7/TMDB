### Question 5

How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?


**Options:**
 - 48
 - 38
 - 14
 - 1

```python
%%sql

SELECT COUNT(DISTINCT m.title) AS thriller_love_movie_count
FROM movies AS m
JOIN genremap AS gp 
ON m.movie_id = gp.movie_id
JOIN keywordmap AS kp 
ON m.movie_id = kp.movie_id
JOIN keywords AS k 
ON kp.keyword_id = k.keyword_id
WHERE gp.genre_id = 53
AND k.keyword_name LIKE '%love%';
```

```python
%%sql

WITH ThrillerMovies AS ( 
	SELECT *
	FROM genres AS g
	INNER JOIN genremap AS gm
	ON g.genre_id = gm.genre_id
	INNER JOIN movies AS m
	ON gm.movie_id = m.movie_id
	WHERE g.genre_name = "Thriller"
),
LoveKeywordMovies AS (
	SELECT *
	FROM keywords AS k
	INNER JOIN keywordmap AS km
	ON k.keyword_id = km.keyword_id
	INNER JOIN movies AS m
	ON km.movie_id = m.movie_id
	WHERE k.keyword_name LIKE '%love%'
)

SELECT 
    COUNT(*)
FROM 
    ThrillerMovies AS tm
INNER JOIN 
    LoveKeywordMovies AS lkm ON tm.movie_id = lkm.movie_id
WHERE CONCAT(' ', lkm.keyword_name, ' ') LIKE '% love %';
```

### Question 8

What are the genres of the movie “The Royal Tenenbaums”?


**Options:**
 - Action, Romance
 - Drama, Comedy
 - Crime, Thriller
 - Drama, Romance

```python
%%sql

SELECT g.genre_name
FROM genres AS g
INNER JOIN genremap AS gm
ON g.genre_id = gm.genre_id
INNER JOIN movies AS m
ON gm.movie_id = m.movie_id
WHERE m.title = "The Royal Tenenbaums";
```

### Question 9

What are the three production companies that have the highest movie popularity score on average, as recorded within the database?


**Options:**

 - MCL Films S.A., Turner Pictures, and George Stevens Productions
 - The Donners' Company, Bulletproof Cupid, and Kinberg Genre
 - Bulletproof Cupid, The Donners' Company, and MCL Films S.A
 - B.Sting Entertainment, Illumination Pictures, and Aztec Musique

```python
%%sql

SELECT  pc.production_company_name,
		AVG(m.popularity) AS average_popularity
FROM productioncompanies AS pc
INNER JOIN productioncompanymap AS pcm
ON pc.production_company_id = pcm.production_company_id
INNER JOIN movies AS m
ON pcm.movie_id = m.movie_id
GROUP BY pc.production_company_name
ORDER BY average_popularity DESC
LIMIT 3;
```

### Question 11

Which genre has, on average, the lowest movie popularity score? 


**Options:**

 - Science Fiction
 - Animation
 - Documentary
 - Foreign

```python
%%sql

SELECT  g.genre_name,
		AVG(m.popularity) AS average_popularity
FROM genres AS g
INNER JOIN genremap AS gm
ON g.genre_id = gm.genre_id
INNER JOIN movies AS m
ON gm.movie_id = m.movie_id
GROUP BY g.genre_name
ORDER BY average_popularity ASC
LIMIT 1;
```

### Question 12

**Options:**

- Special Achievement Award
- Actor in a Supporting Role
- Actress in a Supporting Role
- Best Picture



```python
%%sql

SELECT
    award,
    COUNT(name) AS nomination_count
FROM
    Oscars
WHERE
    award LIKE '%ACTOR%'
GROUP BY
    award
ORDER BY
    nomination_count DESC
```

```python
%%sql

SELECT 	COUNT(name) AS No_of_Nominees,
		award
FROM oscars
GROUP BY award
ORDER BY No_of_Nominees DESC
```
